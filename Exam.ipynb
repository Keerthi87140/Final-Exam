```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Activation, Dense, Dropout, Conv2D, Flatten, MaxPooling2D
from keras.datasets import cifar10  # Import CIFAR-10

# Load CIFAR-10 dataset (if not already loaded)
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Re-define the model if it was lost
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())  # Flatten before Dense layer
model.add(Dense(10, activation='softmax'))  # Assuming 10 classes

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# ... (Your existing code for loading and preprocessing CIFAR-10 data) ...

# Get predictions using CIFAR-10 test data
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = y_test.flatten()  # if y_test is shape (num_samples, 1)

# Class names for CIFAR-10
class_names = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']

# Plot 12 test images with their predicted and true labels
plt.figure(figsize=(12,6))
for i in range(12):
    plt.subplot(3, 4, i+1)
    plt.imshow(x_test[i])
    plt.axis('off')
    true_label = class_names[y_true[i]]
    pred_label = class_names[y_pred_classes[i]]
    color = 'green' if y_true[i] == y_pred_classes[i] else 'red'
    plt.title(f"True: {true_label}\nPred: {pred_label}", color=color, fontsize=9)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# Calculate per-class accuracy
num_classes = 10
per_class_acc = []
for i in range(num_classes):
    idx = np.where(y_true == i)[0]
    acc = accuracy_score(y_true[idx], y_pred_classes[idx])
    per_class_acc.append(acc)

plt.figure(figsize=(8,4))
plt.bar(class_names, per_class_acc, color='skyblue')
plt.ylabel('Accuracy')
plt.xlabel('Class')
plt.title('Per-Class Test Accuracy')
plt.ylim(0,1)
plt.xticks(rotation=45)
plt.show()
```
